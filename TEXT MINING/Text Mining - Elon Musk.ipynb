# AFFINITY SCORES

```python
afinn = pd.read_csv('C:\\Users\\kanis\\Documents\\Assignment\\Afinn.csv', sep=',', encoding='latin-1')
afinn.shape
```

```python
afinn.head()
```

```python
elon_df=pd.DataFrame(elon,columns=['Text'])
```

```python
elon_df
```

```python
affinity_scores = afinn.set_index('word')['value'].to_dict()
```

```python
affinity_scores
```

```python
sentiment_lexicon = affinity_scores
def calculate_sentiment(Text: str = None):
    sent_score = 0
    if Text:
        sentence = nlp(Text)
        for word in sentence:
            sent_score += sentiment_lexicon.get(word.lemma_, 0)
    return sent_score
```

```python
calculate_sentiment(Text = 'amazing')
```

```python
elon_df['sentiment_value'] = elon_df['Text'].apply(calculate_sentiment)
```
